```python
# Get_Tables_List
# Reads pipeline_control and returns active table list

dbutils.widgets.text("catalog_name", "workspace")
dbutils.widgets.text("metadata_schema", "metadata")

catalog_name    = dbutils.widgets.get("catalog_name")
metadata_schema = dbutils.widgets.get("metadata_schema")
```

```python
# Read active tables from pipeline_control
active_tables = spark.sql(f"""
    SELECT table_name
    FROM {catalog_name}.{metadata_schema}.pipeline_control
    WHERE active_flag = 'Y'
    ORDER BY table_name
""").collect()

table_list = [row.table_name for row in active_tables]
print(f"Found {len(table_list)} active tables: {table_list}")
```

```python
# Pass table list to For Each task
dbutils.jobs.taskValues.set(key="table_names", value=table_list)
print("table_names passed to job task values")
```

```python
#verification of table creation

print("Active tables in pipeline_control:")
display(spark.sql(f"SELECT * FROM {catalog_name}.{metadata_schema}.pipeline_control WHERE active_flag = 'Y' ORDER BY table_name"))
```
